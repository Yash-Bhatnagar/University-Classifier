### Model Creation

```python
from sklearn.cluster import KMeans
```

```python
# Number of clusters = 2 , 1 each for Public and Private
kmeans = KMeans(n_clusters=2)
```

```python
X = df.drop(columns='Private')
kmeans.fit(X)
```

```python
# The 2 cluster centers
print("Cluster Centers")
print("\nCluster Center 1 : ",kmeans.cluster_centers_[0][:])
print("\nCluster Center 2 : ",kmeans.cluster_centers_[1][:])
```

```python
# Predicted labels
kmeans.labels_
```

## Evaluation

There is no perfect way to evaluate clustering if you don't have the labels.

But we do have the labels, so we take advantage of this to evaluate our clusters, this won't be usually available.

**Let's convert 'Cluster', which is a 1 for a Private school, and a 0 for a public school.**

```python
df['Private'] = df['Private'].apply(lambda x: 1 if x == 'Yes' else 0)
```

```python
df['Private'].value_counts()
```

```python
from sklearn.metrics import confusion_matrix,classification_report

print(confusion_matrix(df['Private'],kmeans.labels_))
```

```python
print(classification_report(df['Private'],kmeans.labels_))
```

For an unsupervised model, this seems to be a pretty good score.

**Note: In in real world situation we won't have anything to compare from.**
